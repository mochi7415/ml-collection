### 数理モデル 
- 数学的表現を用いて現実世界のシステムやプロセスを模倣、説明、予測するモデル
- 一般に、方程式、不等式、論理構造などの数式を使って表現


#### BMIについて

BMI（Body Mass Index、体格指数）は、人の体重（キログラム）を身長（メートル）の二乗で割った数理モデル
$$ BMI=\frac{体重(kg)}{身長(m)^2} $$

一般的なBMI基準
- 18.5未満：低体重（痩せ型）
- 18.5～24.9：正常体重
- 25.0～29.9：過体重（前肥満）
- 30.0以上：肥満（1度、2度、3度の肥満に細分化されることもあります）


```python
# BMIを計算する

def calculate_bmi(weight, height):
    bmi = weight / (height ** 2)
    return bmi

# 体重と身長を入力してBMIを計算する
height = 1.75  # 身長（メートル）
weight = 70    # 体重（キログラム）

bmi = calculate_bmi(weight, height)
print(f"BMI: {bmi:.2f}")
```

```python
# 健康評価

def evaluate_bmi(height, weight):
    height = height / 100  # 身長をセンチメートルからメートルに変換
    bmi = weight / (height ** 2)  # BMIの計算式

    # BMIに基づいて健康状態を判定
    if bmi < 18.5:
        return "低体重"
    elif 18.5 <= bmi < 25:
        return "正常体重"
    elif 25 <= bmi < 30:
        return "過体重"
    else:
        return "肥満"

# 使用例:
height = 175  # 身長（センチメートル）
weight = 70  # 体重（キログラム）

status = evaluate_bmi(height, weight)
print(status)
```

```python
# ランダムフォレストを使用してBMIの健康評価を行う


# BMIのデータをDataFrameにまとめる

import pandas as pd

# データの宣言
data = {
    '身長': [174.6, 152.0, 157.9, 178.4, 156.6],
    '体重': [79.9, 49.2, 54.1, 103.1, 54.0],
    '判定': ['過体重', '正常体重', '正常体重', '肥満', '正常体重']
}

# DataFrameの作成
df = pd.DataFrame(data)

# 表示
print(df)

x = df[['身長', '体重']] # 説明変数
y = df['判定']  # 目的変数

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
# データの分割
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# ランダムフォレストモデルの作成
model = RandomForestClassifier(n_estimators=100, random_state=42)
# モデルの訓練
model.fit(x_train, y_train)
# モデルの評価
accuracy = model.score(x_test, y_test)
print(f"\nAccuracy: {accuracy:.2f} \n")


def predict_health(height, weight):
    # 特徴名を使ってDataFrameを作成
    data = pd.DataFrame([[height, weight]], columns=['身長', '体重'])
    # モデルで予測
    pred = model.predict(data)
    return pred[0]

# テスト
height = 175  # 身長（センチメートル）
weight = 50   # 体重（キログラム）
health_status = predict_health(height, weight)
print(f"予測された健康状態: {health_status}")


```

より性能高いモデルを作るためには、より多くのデータが必要。

また、モデルの性能を改善するためには、特徴量エンジニアリング、ハイパーパラメータのチューニング、さらに複雑なモデルの試行など工夫が必要

### 数理モデルと機械学習の違い
数理モデルと機械学習は互いに補完的な関係

1.数理モデル
- 数理モデルは、現象やプロセスを数学的な言語で表現したものです。これらのモデルは、システムの挙動を理解し、予測するために、既知の理論や法則に基づいて構築されます。数理モデルの構築には、システムの基本的なメカニズムに関する深い理解が必要であり、モデルの精度はその理論的基礎に大きく依存します。

2.機械学習
- 機械学習は、データからパターンを学習し、その学習を通じて予測や決定を行うアルゴリズムの集合です。機械学習モデルは、特定の理論的背景に依存することなく、大量のデータから直接パターンを抽出することを目的としています。このアプローチの主な利点は、データ駆動型であり、未知の関係や複雑なシステムでも効果的に機能する可能性があることです。


### 機械学習が注目集める理由

1.大量データの利用可能性
- デジタル化の進展により、膨大な量のデータが生成され、収集されるようになりました。機械学習は、このようなビッグデータからパターンを学習し、予測モデルを構築するのに特に適しています。

2.計算能力の向上
- GPUの発展やクラウドコンピューティングサービスの普及により、以前は不可能だった複雑な機械学習モデルの訓練が可能になりました。これにより、より高度な分析が実現可能となっています。

3.数理モデルでは困難な問題の解決
- 特定の問題に対して、数理モデルでは表現や解析が難しい複雑な関係や非線形なパターンを機械学習モデルは捉えることができます。これにより、より幅広い問題に対処することが可能になりました。

4.新たな発見と洞察の提供
- 機械学習は、人間では識別が難しい複雑なパターンや関係を識別する能力を持っています。新たな発見や理解の深化につながります。


### 数理モデルで解決困難な問題

数理モデルでは解決が困難な問題は、主にデータの複雑性、非線形性、および高次元性に関連している

1. 顔認識

顔認識は、個人の顔の特徴を識別して特定の人物を認識するプロセスです。顔のデータは非常に複雑で、人によって異なる多様な特徴（目の形、鼻の形、表情など）を含んでいます。また、照明、角度、表情の変化によってもその見え方は大きく変わります。これらの要素を厳密な数理モデルで表現するのは非常に困難ですが、機械学習を用いることで、大量の顔画像データから有用なパターンを学習し、高い精度で顔認識を実現できます。

2. 自然言語処理（NLP）

自然言語処理は、人間の言語をコンピュータで理解し、処理する技術です。言語は非常に複雑で、文脈、文法、意味論、語彙の多様性など、数理モデルで完全に捉えることが困難な要素が多く含まれています。例えば、同じ単語が異なる文脈で全く異なる意味を持つことがあります。機械学習、特に深層学習を用いると、大量のテキストデータからこれらの複雑な言語のパターンを学習し、翻訳、感情分析、質問応答などが実現するようになりました。

3. 推薦システム

推薦システムは、ユーザーの過去の行動や好みに基づいて、商品やコンテンツを推薦する技術です。ユーザーの好みは多様で、時間とともに変化します。また、商品やコンテンツの関連性は非線形であり、ユーザー間の相互作用や社会的影響を考慮する必要があります。これらを厳密な数理モデルで表現するのは非常に困難ですが、機械学習技術を用いることで、ユーザーデータから複雑な嗜好パターンを抽出し、個々のユーザーに合わせた精度の高い推薦を提供できます。
