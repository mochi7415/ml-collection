### 行列乗算


行列乗算が応用される代表的なアプリケーション

行列乗算は、科学、工学、経済学、社会科学など、多岐にわたる分野で基礎的かつ重要な役割

- コンピュータグラフィックス: 3Dモデリングやアニメーションで、物体の回転、拡大縮小、移動などを行うために行列乗算が用いられます。これにより、現実的なビジュアルエフェクトの作成が可能になります。

- 機械学習とデータ分析: 多くの機械学習アルゴリズム、特にニューラルネットワークは、重みと入力データ間の行列積を計算することで機能します。これにより、予測、分類、データのパターン認識が可能になります。

- 物理学とエンジニアリング: 力学、電磁気学、量子力学など、多くの物理理論は行列を使って表されます。行列乗算により、システムの状態変化や相互作用をモデル化できます。

- 経済学: 経済モデルで市場の平衡状態を解析する際や、リスク管理でポートフォリオの最適化を行う際に行列乗算が使用されます。

- ロボティクス: ロボットの運動学において、ロボットの各部位の位置、速度、加速度を計算するために行列乗算が用いられます。これにより、正確な動きとタスクの実行が可能になります。

- 信号処理: 画像処理、音声処理、通信システムにおいて、フィルタリング、信号の変換、ノイズの削減などを行うために行列乗算が使用されます。

- 統計学: 多変量解析や主成分分析（PCA）など、データセット間の関係を解析するために行列積が用いられます。

- 地質学と地理情報システム（GIS）: 地球の表面や地下構造のモデリング、空間データの解析などに行列乗算が利用されます。

- 生物情報学: ゲノム解析やタンパク質の構造予測において、大量の生物学的データを処理し、パターンや関係性を見つけ出すために行列乗算が使われます。

- 金融工学: 資産価格のモデル化、リスク評価、デリバティブの価格設定などに行列乗算が広く使用されます。


$$
\begin{bmatrix}
       &        &        &        \\
       &        &        &        \\
a_{i1} & a_{i2} & \dots  & a_{im} \\
       &        &        &        \\
\end{bmatrix}
\begin{bmatrix}
       &        & b_{1j} &        \\
       &        & b_{2j} &        \\
       &        & \vdots &        \\
       &        & b_{mj} &        \\
\end{bmatrix}
=
\begin{bmatrix}
       &        & \vdots &        \\
       &        & \vdots &        \\
 \dots & \dots  & c_{ij} & \dots  \\
       &        & \vdots &        \\
\end{bmatrix}
$$


$$ c_{ij} = \sum_{k=1}^{n}a_{ik}b_{kj}$$

$$
A=
\begin{bmatrix}
 a_{11} & a_{12} \\
 a_{21} & a_{22} \\
\end{bmatrix}
,
B=
\begin{bmatrix}
 b_{11} & b_{12} \\
 b_{21} & b_{22} \\
\end{bmatrix}
$$

$$
C=AB=
\begin{bmatrix}
 a_{11}b_{11}+a_{12}b_{21} & a_{11}b_{12}+a_{12}b_{22} \\
 a_{21}b_{11}+a_{22}b_{21} & a_{21}b_{12}+a_{22}b_{22} \\
\end{bmatrix}
$$







```python
# 行列乗算
import random

def random_matrix(n):
    Z = []
    for i in range(n):
        # 新しい行を作成
        row = []
        for j in range(n):
            # 各行に乱数を追加
            row.append(random.randint(0,9))
        # 完成した行を行列に追加
        Z.append(row)
    return Z


def print_matrix(matrix):
    for row in matrix:
        print(" ".join("{:4}".format(x) for x in row))



def matmul(X, Y):
    # 行列のサイズを決定する
    n = len(X)

    # 結果を格納する行列をゼロで初期化する
    Z = [[0 for _ in range(n)] for _ in range(n)]

    # 行列積を計算する
    for i in range(n): # すべての行に対して
        for j in range(n): # すべての列に対して
            for k in range(n): # 乗算と加算を実行する
                Z[i][j] += X[i][k] * Y[k][j]

    return Z


A= random_matrix(5)  # 5x5行列を作る
print("行列A:")
print_matrix(A)
B = random_matrix(5)  # 5x5行列を作る
print("\n行列B:")
print_matrix(B)
C = matmul(A, B)     # AとBの行列積を計算する
print("\n行列AとBの積:")
print_matrix(C)
```

##### 行列の大きさと実行時間

実際どのくらいの大きさの行列が使われているのか

1. コンピュータグラフィックス
応用例: 3Dオブジェクトの変換（回転、スケーリング、平行移動）に使用されます。
行列のサイズ: 変換には`4x4`の行列がよく使われます。
2. 機械学習とディープラーニング
応用例: ニューラルネットワークでは、重みと入力データの行列積が計算され、特徴を変換します。
行列のサイズ: 入力層や隠れ層における行列は、`数百から数千`の要素を含むことが一般的です。
3. シミュレーションと数値解析
応用例: 物理シミュレーションや工学解析で、状態や力の分布を計算します。
行列のサイズ: システムの複雑さに応じて、`数千から数百万`の要素を持つ行列が用いられます。
4. データサイエンス
応用例: データセット間の関係を探るため、共分散行列や相関行列の計算に行列積が用いられます。
行列のサイズ: データの特性や量に応じて、`数十から数百万`の要素を持つ行列が使用されます。
5. 自然言語処理（NLP）
応用例: 単語の埋め込み表現や文脈を考慮したモデリングに行列積が用いられます。
行列のサイズ: 単語の埋め込み表現では、`数百`の要素を持つ行列が一般的です。


```python
# nを大きくしたときの実行時間

import matplotlib.pyplot as plt
import time

# 行列乗算
import random

def random_matrix(n):
    Z = []
    for i in range(n):
        # 新しい行を作成
        row = []
        for j in range(n):
            # 各行に乱数を追加
            row.append(random.randint(0,9))
        # 完成した行を行列に追加
        Z.append(row)
    return Z


def print_matrix(matrix):
    for row in matrix:
        print(" ".join("{:4}".format(x) for x in row))



def matmul(X, Y):
    # 行列のサイズを決定する
    n = len(X)

    # 結果を格納する行列をゼロで初期化する
    Z = [[0 for _ in range(n)] for _ in range(n)]

    # 行列積を計算する
    for i in range(n): # すべての行に対して
        for j in range(n): # すべての列に対して
            for k in range(n): # 乗算と加算を実行する
                Z[i][j] += X[i][k] * Y[k][j]

    return Z


# nの値と実行時間を記録する
n_values = [10, 20, 50, 100, 200]
times = []
for n in n_values:
    A = random_matrix(n)
    B = random_matrix(n)
    start_time = time.time()
    C = matmul(A, B)
    times.append(time.time() - start_time)

# 結果をプロットする
plt.plot(n_values, times, marker='o')
plt.xlabel('size of matrix (n x n)')
plt.ylabel('execution time (seconds)')

```

### GPUを用いた並列計算

- 行列乗算はGPUの並列性を活かす典型的な例で、各要素の計算を異なるスレッドに割り当てることで、全体の計算時間を大幅に短縮可能
- 深層学習用のライブラリ PyTorchを使うことで、GPUを活用した計算が活用できる

```python
# PyTorchで行列乗算の実行時間を計測する
# GPUを利用した行列乗算の実装

import torch
import time
import numpy as np

def random_matrix(size):
    return np.random.rand(size, size)

def time_matmul_torch(n, iteration=10):
    # NumPyの行列を生成
    A_np = random_matrix(n)
    B_np = random_matrix(n)

    # NumPyの行列をPyTorchのテンソルに変換
    A_torch = torch.from_numpy(A_np).float()
    B_torch = torch.from_numpy(B_np).float()

    # GPUが利用可能ならテンソルをGPUに移動
    if torch.cuda.is_available():
        A_torch = A_torch.cuda()
        B_torch = B_torch.cuda()
        print("Using GPU")
    else:
        print("GPU is not available, using CPU")

    # 計算時間を測定
    times = []
    for _ in range(iteration):
        start_time = time.time()
        torch.mm(A_torch, B_torch)  # 行列A, Bの乗算
        if torch.cuda.is_available():
            torch.cuda.synchronize()  # GPUの全てのキューが完了するまで待機
        end_time = time.time()
        times.append(round(end_time - start_time, 3))

    print(f'PyTorch n={n}', times)
    return round(sum(times) / iteration, 3)

# テスト
n = 2000  # 行列サイズ
iteration = 10  # 繰り返し回数
avg_time = time_matmul_torch(n, iteration)
print(f"Average time: {avg_time} seconds")
```
